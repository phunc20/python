```python
class Foot:
    def __init__(self, length, width):
        # in cm
        self.l = length
        self.w = width
    def __repr__(self):
        return f"Foot: length={self.l}, width={self.w}"
```

```python
import random
n_feet = 1_000
L = []
for _ in range(n_feet):
    l, w = random.randint(15, 45), random.randint(5, 20)
    #print(l, w)
    L.append(Foot(l ,w))
L
```

```python
from copy import deepcopy, copy
```

```python
%timeit L1 = deepcopy(L)
```

```python
%timeit L1 = copy(L)
```

```python
%timeit L1 = L
```

```python
def swap(LL):
    LL[0], LL[1] = LL[1], LL[0]

L2 = [3.14, 2.71]
swap(L2)
L2
```

```python
def swap_return(LL):
    LL[0], LL[1] = LL[1], LL[0]
    return LL
L3 = [1.1, 2.2]
L4 = swap_return(L3)
L3, L4
```

```python
L3 is L4
```

```python
%timeit swap(L2)
```

```python
%timeit swap_return(L2)
```
