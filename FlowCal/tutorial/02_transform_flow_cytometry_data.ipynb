```python
import FlowCal
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("dark_background")
```

```python
s = FlowCal.io.FCSData("sample001.fcs")
```

```python
plt.hist(s[:, "FL1"], bins=100);
```

```python
s_transformed = FlowCal.transform.to_rfi(
    s, channels="FL1")

bins = np.logspace(0, 4, 100)
plt.hist(s_transformed[:, "FL1"], bins=bins)
plt.xscale("log")
```

- `FlowCal.transform.to_rfi(s)` transforms all channels
- `FlowCal.transform.to_rfi(s, channels="FL1")`
- `FlowCal.transform.to_rfi(s, channels=["FL1", "FL2", "FL3"])`
