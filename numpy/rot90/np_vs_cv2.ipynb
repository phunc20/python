```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = plt.imread("mona_lisa.jpg")
plt.imshow(image);
```

```python
fig, ax = plt.subplots(1,2)
ax[0].imshow(np.rot90(image))
ax[1].imshow(cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE));
```

```python
%%timeit
cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
```

```python
%%timeit
np.rot90(image)
```

```python
%%timeit
np.flip(np.transpose(image, (1,0,2)), axis=0)
```

```python
plt.imshow(np.flip(np.transpose(image, axes=[1,0,2]), axis=0));
```

```python
np.rot90(image) is image
```

```python
cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE) is image
```

```python
np.rot90(image)[-1,0] is image[0,0]
```

```python
np.array_equal(np.rot90(image)[-1,0], image[0,0])
```

```python
np.rot90(image)[-1,0,0] is image[0,0,0]
```

```python
%%timeit
cv2.rotate(image, cv2.ROTATE_180)
```

```python
%%timeit
np.rot90(image, k=2)
```

```python
np.flip(image) is image
```

```python
np.flip(image)[0,-1] is image[0,0]
```

```python
np.array_equal(np.flip(image)[0,-1], image[0,0])
```

```python
plt.imshow(np.flip(image));
```

```python
plt.imshow(np.flip(image, axis=0));
```

```python
np.flip(image, axis=1)[0, 0, 0] is image[0, -1, 0]
```

```python
np.array_equal(np.flip(image, axis=1)[0, 0], image[0, -1])
```
