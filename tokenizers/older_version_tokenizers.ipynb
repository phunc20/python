## Python Version (Unnecessary)

Some said that one might need to downgrade Python version in order to install older versions of `tokenizers`. However, in our experience, as of 2024/12, we seem to be able to keep the default Python of Colab.

In case you must, here below seems to be one way to downgrade.

[Python versions and their release dates](https://www.python.org/doc/versions/)

```python
!apt search "^python3"
```

Although we did not find Python 3.7 during the search, we can still install it. Hay chưa?

```python
!apt install python3.7
```

Simply installing like this won't make the `python` command version 3.7, as the following cell demonstrates.

```python
!python --version
```

```python
!which python3.7
```

To make 3.7 the default Python, we need another Ubuntu command, i.e. `update-alternatives`:

```python
#update-alternatives --install <link> <name> <path> <priority>
!update-alternatives --install /usr/bin/python3 python3 /usr/bin/python3.7 1
```

```python
!python --version
```

```python
!update-alternatives --config python3
```

```python
!apt install python3-pip
```

```python
!apt install python3.7-distutils
```

## Rust to 2021/10 Version or Before

- <https://releases.rs/docs/1.54.0/> (2021/07/29)
- ...
- <https://releases.rs/docs/1.50.0/> (2021/02/11)

We need the right version of Rust because the `tokenizers` package depends on Rust.

```python
# Check whether Rust is installed and remove it if installed
!which rustc && rustc --version && apt remove rustc
```

Let's install `rustup` instead, which seems to facilitate switching versions later on.

```python
!curl --proto '=https' --tlsv1.2 -sSf https://sh.rustup.rs | sh -s -- -y
```

This installed a bunch of commands/binaries under `~/.cargo/bin` but our `PATH` environment variable is not automatically updated.

```python
!ls ~/.cargo/bin
```

```python
!echo $PATH | grep cargo
```

Let's put it in `PATH` then.

```python
import os
```

```python
os.environ["PATH"] = f"{os.environ['HOME']}/.cargo/bin:{os.environ['PATH']}"
```

Let's do a double-check.

```python
!echo $PATH | cut -d: -f1
```

Great! Next configure Rust to a version close to the release time of `tokenizers==0.10.3` (i.e. 2021/05/25)
- <https://pypi.org/project/tokenizers/#history>
- <https://releases.rs/docs/1.50.0/>

**How do you know that you should install Rust version `1.50.0`**?

Well, I googled and debugged it. I seemed to only lose the exact link, though. The best I could recover from my search is the following link:
<https://github.com/rust-lang/rust/issues/81654>

> Installing `tokenizers==0.10.3` using `rustc==1.83.0` (i.e. the stable version at the time of writing)
> will give rise to error messages which might also
> shed some light on how to debug this.

```python
rustc_version = "1.50.0"
```

```python
!echo {rustc_version}
```

```python
!rustup default {rustc_version}
```

Ok. Let's check whether Rust's version has been correctly configured.

```python
!which rustc
```

```python
!rustc --version
```
